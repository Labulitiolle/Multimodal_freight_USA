# Test Roads

```python
from mfreight.Road.gen_road_net import RoadNet
from shapely.geometry import Polygon

G_road, nodes_road, edges_road = RoadNet(
                                        bbox= Polygon([(-88, 24), (-88, 31), (-79, 31), (-79, 24)])).gen_road_graph(
                                                                                                      simplified=True,
                                                                                                      save=True,
                                                                                                      return_gdfs=True)
```

```python
import osmnx as ox
fig, ax = ox.plot_graph(G_road,node_color='blue',bgcolor='white',node_alpha=0.5)
```
