```python
web = pd.read_html('https://www.bts.gov/archive/subject_areas/freight_transportation/faf/users_guide/')
```

```python
dms_codes = web[2]
```

```python
region_map = dms_codes.set_index('Code').loc[:,'FAF Region*']
state_map = dms_codes.set_index('Code').loc[:,'State']
```

```python
df3['state_orig'] = df3['dms_orig'].replace(state_map)
df3['state_dest'] = df3['dms_dest'].replace(state_map)

df3['region_orig'] = df3['dms_orig'].replace(region_map)
df3['region_dest'] = df3['dms_dest'].replace(region_map)
```
