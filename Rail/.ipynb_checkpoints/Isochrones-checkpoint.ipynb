## Get closest intermodal node from door

```python
shp_nodes_relevant_cleaned = gpd.read_file("rail_data/cleaned_data/rail_nodes_cleaned/shp_nodes_relevant_cleaned.shp")
intermodal_nodes = shp_nodes_relevant_cleaned[shp_nodes_relevant_cleaned.FRANODEID < 300].copy()
intermodal_nodes.head(1)
```

```python
def lat_long_intermodal_map(row):
    row['lat_long'] = row['geometry'].coords[0][::-1]
    return row

intermodal_nodes_coords = intermodal_nodes.apply(lambda x: lat_long_intermodal_map(x), axis=1)
intermodal_map = intermodal_nodes_coords.set_index("FRANODEID").lat_long
intermodal_map
```

```python
def dist_to_intermodal(value, point):
    return great_circle(point,value).km
    

def find_closest_intermodal_node(position, intermodal_map):
    intermodal = intermodal_map.copy().apply(dist_to_intermodal, args=(position,))
    
    return intermodal.idxmin()
```

```python
pt = (45,-123)
closest_node = find_closest_intermodal_node(pt,intermodal_map)
closest_node
```

```python
def get_closest_node_lat_lon(pt,intermodal_map):
    closest_node = find_closest_intermodal_node(pt,intermodal_map)
    return intermodal_map[closest_node]
```
