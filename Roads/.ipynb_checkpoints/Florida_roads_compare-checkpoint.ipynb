# Comparing the routing performances of the road graphs 

```python
import networkx as nx
import osmnx as ox

import matplotlib.pyplot as plt

import re
import pandas as pd
import geopandas as gpd

pd.set_option('display.max_columns', None)
```

```python
G_BTS = nx.read_gpickle('data/BTS_florida_G.plk')

G_OSM = nx.read_gpickle('data/OSM_florida_G.plk')
```

```python
def sum_list(row):
    if isinstance(row['duration'],list):
        row['duration'] = sum(row['duration'])
    else:
        pass
    return row
```

```python
def add_edge_time_graph(df):
    df = df.apply(lambda x: sum_list(x),axis=1)
    return df

def clean_graph(G):
    nodes, edges = ox.graph_to_gdfs(G)
    edges = add_edge_time_graph(edges)
    
    edges = edges.drop(edges.columns.difference(['length','duration','geometry','u', 'v', 'key']), axis=1)
    G = ox.graph_from_gdfs(nodes, edges)
    return G
```

```python
nodes, edges = ox.graph_to_gdfs(G_OSM)
print(edges.columns)
```

```python
G_BTS = clean_graph(G_BTS)
G_OSM = clean_graph(G_OSM)
```

```python
def get_edge_attribute(G, node=0):
    i,j = list(G.edges())[node]
    return G.edges[i,j,0]
```

```python
fig, ax = ox.plot_graph(G_BTS,node_color='blue',bgcolor='white',node_alpha=0.5)
```
